```python
def clean_descriptions(x):
    new = ""
    if x == "a residential postcode":
        new = "R"
    elif x == "a non-residential postcode":
        new = "NR"
    elif x =="a mixed residential and non-residential postcode":
        new = "M"
    elif x == "a primarily residential postcode":
        new = "PR"
    elif x == "a primarily non-residential postcode":
        new = "PNR"
    elif x ==  "a large user postcode":
        new = "LU"
    elif x == "a defunct large user postcode":
        new = "LU"
    else:
        new = "U"
    return new

```

```python
bike_locations["description_code"]=bike_locations["description"].apply(clean_descriptions)
```

```python
#setting up a loop to find the nearest bus stops
import time
bike_locations["nearest_bus"]=""
print (bike_locations.shape[0])
for i in range(bike_locations.shape[0]):
    lat=str(bike_locations.at[i, "latitude_x"])
    lon=str(bike_locations.at[i, "longitude_x"])
    url="https://api.tfl.gov.uk/StopPoint?radius=200&stopTypes=NaptanBusCoachStation,NaptanPublicBusCoachTram&lat="+lat+"&lon="+lon+"&app_key="+app_key
    response = requests.get(url)
    try:
        bike_locations.at[i, "nearest_bus"]=response.json()["stopPoints"][0]["distance"]
    except IndexError:
        url="https://api.tfl.gov.uk/StopPoint?radius=500&stopTypes=NaptanBusCoachStation,NaptanPublicBusCoachTram&lat="+lat+"&lon="+lon+"&app_key="+app_key
        response = requests.get(url)
        try:
            bike_locations.at[i, "nearest_bus"]=response.json()["stopPoints"][0]["distance"]
        except IndexError:
            bike_locations.at[i, "nearest_bus"]="further than 500m"
    print (i, response.status_code)
    time.sleep(0.5)
```

```python
#loop to find the nearest tube stations
bike_locations["nearest_tube"]=""
print (bike_locations.shape[0])
for i in range(bike_locations.shape[0]):
    lat=str(bike_locations.at[i, "latitude_x"])
    lon=str(bike_locations.at[i, "longitude_x"])
    url="https://api.tfl.gov.uk/StopPoint?radius=500&stopTypes=NaptanMetroStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
    response = requests.get(url)
    try:
        bike_locations.at[i, "nearest_tube"]=response.json()["stopPoints"][0]["distance"]
    except IndexError:
        url="https://api.tfl.gov.uk/StopPoint?radius=1000&stopTypes=NaptanMetroStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
        response = requests.get(url)
        try:
            bike_locations.at[i, "nearest_tube"]=response.json()["stopPoints"][0]["distance"]
        except IndexError:
            bike_locations.at[i, "nearest_tube"]="further than 1km"
    print (i, response.status_code)
    time.sleep(0.4)
```

```python
# loop to find nearest railway station
bike_locations["nearest_rail"]=""

print (bike_locations.shape[0])
for i in range(bike_locations.shape[0]):
    lat=str(bike_locations.at[i, "latitude_x"])
    lon=str(bike_locations.at[i, "longitude_x"])
    url="https://api.tfl.gov.uk/StopPoint?radius=500&stopTypes=NaptanRailStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
    response = requests.get(url)
    try:
        bike_locations.at[i, "nearest_rail"]=response.json()["stopPoints"][0]["distance"]
    except IndexError:
        url="https://api.tfl.gov.uk/StopPoint?radius=1000&stopTypes=NaptanRailStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
        response = requests.get(url)
        try:
            bike_locations.at[i, "nearest_rail"]=response.json()["stopPoints"][0]["distance"]
        except IndexError:
            bike_locations.at[i, "nearest_rail"]="further than 1km"
    print (i, response.status_code)
    time.sleep(0.4)
```

```python
bike_locations['nearest_rail']=bike_locations['nearest_rail'].astype(float) 
```

*rerunning loop for docks that were too far away from a station. The larger the search radius the slower the request hence running it twice*

```python
#adding rail station when the nearest was further than 1km away with increased search radius
for i in range(bike_locations.shape[0]):
    
    if bike_locations.at[i, "nearest_rail"]=="further than 1km":
        lat=str(bike_locations.at[i, "latitude_x"])
        lon=str(bike_locations.at[i, "longitude_x"])
        url="https://api.tfl.gov.uk/StopPoint?radius=2000&stopTypes=NaptanRailStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
        response = requests.get(url)
        try:
            bike_locations.at[i, "nearest_rail"]=response.json()["stopPoints"][0]["distance"]
        except IndexError:
            url="https://api.tfl.gov.uk/StopPoint?radius=3000&stopTypes=NaptanRailStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
            response = requests.get(url)
            try:
                bike_locations.at[i, "nearest_rail"]=response.json()["stopPoints"][0]["distance"]
            except IndexError:
                bike_locations.at[i, "nearest_rail"]="further than 3km"
        print (i, response.status_code)
        time.sleep(0.4)
    else:continue
```

```python
# adding nearest tube stations for docks where the nearest tube is further than 1km away with increased search radius
print (bike_locations.shape[0])
for i in range(bike_locations.shape[0]):
    if bike_locations.at[i, "nearest_tube"]=="further than 1km":
        lat=str(bike_locations.at[i, "latitude_x"])
        lon=str(bike_locations.at[i, "longitude_x"])
        url="https://api.tfl.gov.uk/StopPoint?radius=1500&stopTypes=NaptanMetroStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
        response = requests.get(url)
        try:
            bike_locations.at[i, "nearest_tube"]=response.json()["stopPoints"][0]["distance"]
        except IndexError:
            url="https://api.tfl.gov.uk/StopPoint?radius=3000&stopTypes=NaptanMetroStation&lat="+lat+"&lon="+lon+"&app_key="+app_key
            response = requests.get(url)
            try:
                bike_locations.at[i, "nearest_tube"]=response.json()["stopPoints"][0]["distance"]
            except IndexError:
                bike_locations.at[i, "nearest_tube"]="further than 3km"
        print (i, response.status_code)
        time.sleep(0.4)
```
